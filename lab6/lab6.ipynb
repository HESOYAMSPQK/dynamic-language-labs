```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('nba.csv') # Загрузка данных из CSV-файла

data.head(5)
```

## Matplotlib:


### 1.1. Пример построения графика с использованием matplotlib:

```python
import matplotlib.pyplot as plt

# Построение графика возраста игроков
plt.plot(data['Age'])
plt.xlabel('Номер')
plt.ylabel('Возраст')
plt.title('График возраста игроков NBA')
plt.show()
```

### 1.2. Пример построения нескольких суб-графиков на одном графике:

```python
# Создание суб-графиков
fig, axs = plt.subplots(2, 2)

# График возраста
axs[0, 0].plot(data['Age'])
axs[0, 0].set_title('Возраст')

# График роста
axs[0, 1].plot(data['Weight'])
axs[0, 1].set_title('Вес')

# График зарплаты
axs[1, 0].plot(data['Salary'])
axs[1, 0].set_title('Зарплата')

# График номера игрока
axs[1, 1].plot(data['Number'])
axs[1, 1].set_title('Номер игрока')

# Установка общего заголовка
fig.suptitle('Суб-графики различных параметров игроков NBA')
plt.show()
```

## 2.Графики с использованием pandas.DataFrame.plot / Series.plot:

```python
# График роста с использованием DataFrame.plot
data.plot(x='Name', y='Weight', kind='bar', rot=90, title='График роста игроков')

# График зарплаты с использованием Series.plot
data['Salary'].plot(kind='hist', title='Гистограмма зарплат игроков')

# Boxplot для зарплаты в разрезе команды
data.boxplot(column='Salary', by='Team', figsize=(10, 6))
plt.title('Boxplot зарплаты игроков по командам')
plt.xticks(rotation=90)
plt.show()
```
